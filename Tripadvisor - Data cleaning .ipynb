```python
#some page names not aligned (name vs. page name scraped from separate urls)
df1 = destinations[['Name', 'Featured Price', 'Category']]
df2 = destinations[['Page Name', 'Tripadvisor Rating',
       'Address', 'Property Amenities', 'Room Features', 'Room Types',
       'Hotel Style', 'Languages', 'Hotel Star Class', 'Car/Walkability',
       'Restaurants Proximity', 'Attractions Proximity', 'Number of Reviews',
       'Covid Precautions', 'Travel Destination']]
```

```python
df2['Name'] = df2['Page Name']
```

```python
full_df = df1.merge(df2, how='left', on = 'Name')
```

```python
full_df
```

```python
full_df = full_df.drop_duplicates(subset=['Name', 'Address'])
```

```python
full_df.shape
```

```python
full_df
```

```python
#remove the sponsored pages' rows
full_df = full_df[~(full_df['Name'] != full_df['Page Name'])]
```

```python
full_df.shape
```

```python
#remove rows with empty Tripadvisor Ratings
full_df = full_df[~(full_df['Tripadvisor Rating'] == 'empty')]
```

```python
#full_df['Featured Price'].unique()
#clean price
def clean_price(price):
    if price != 'empty':
        price = price.split()
        price = float((price[0][1:]))
    return price
    
```

```python
full_df
```

```python
full_df['Featured Price'] = full_df['Featured Price'].apply(clean_price)
```

```python
full_df['Featured Price']
```

```python
#hotel star class
#full_df['Hotel Star Class'].unique()
def clean_class(stars):
    if stars != 'empty':
        stars = stars.split(' ')
        stars = float(stars[0])
    return stars
```

```python
full_df['Hotel Star Class'] = full_df['Hotel Star Class'].apply(clean_class)
```

```python
full_df.info()
```

```python
#tripadvisor rating into floats
full_df['Tripadvisor Rating'] = full_df['Tripadvisor Rating'].apply(lambda x: float(x))
```

```python
#number of reviews-none missing since only kept rows with tripadvisor ratings (connected)
full_df['Number of Reviews'] = full_df['Number of Reviews'].apply(lambda x: int(x.split(' ')[0].replace(',', '')))
```

```python
full_df.head()
```

```python
#clean Travel Destination 
full_df['Travel Destination'] = full_df['Travel Destination'].apply(lambda x: x.split('.')[0])
```

```python
#full_df['Restaurants Proximity'].unique()
#full_df['Attractions Proximity'].unique()

def clean_proximity(prox):
    if prox != 'empty':
        prox = prox.split()
        prox = round((float(prox[0]) / float(prox[3])), 2)
    return prox

#left emptys for now, but can impute average based on location or leave out depending on how many
```

```python
full_df['Restaurants/Mile'] = full_df['Restaurants Proximity'].apply(clean_proximity)

full_df['Attractions/Mile'] = full_df['Attractions Proximity'].apply(clean_proximity)
```

```python
full_df.info()
```

```python
full_df[full_df['Car/Walkability'] == 'empty']
#interesting that 305 rows missing for car/walkability, restaurants, attraction - remote hotels? impute based on zip code?
```

```python
#drop the 7 rows with empty address - looks to be scraping error after manual search
#full_df[full_df['Address'] == 'empty']
full_df = full_df[~(full_df['Address'] == 'empty')]
```

```python
#zip code column 

def make_zip(address):
    zip_code = address.split(',')[-1].strip()
    return zip_code
```

```python
full_df['Zip Code'] = full_df['Address'].apply(make_zip)
```

```python
#clean car/walkability - walkability scores below 50 are car-recommended
def clean_walkability(score):
    if score != 'empty':
        score = score.split(' ')[-4]
    return score
```

```python
full_df['Walk Score'] = full_df['Car/Walkability'].apply(clean_walkability)
```
